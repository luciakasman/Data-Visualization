```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
datos = pd.read_csv('events.csv')
datos
```

```python
paises = datos[["country", "person"]].dropna().drop_duplicates("person")
paises = paises.groupby("country")
paises = paises.count().reset_index().sort_values("person", ascending=False)
paises.count()
```

```python
paises["person in log"] = np.log(paises["person"]+1)
```

```python
ax = sns.barplot(x=paises['country'], y=paises['person in log'],edgecolor=(0,0,0), linewidth=2)
ax.set_title('Cantidad de personas vs. pases desde donde entran a la web',fontsize=20)
ax.set_ylabel('Cantidad de personas (log)',fontsize=14)
ax.set_xlabel('Pases',fontsize=14)
ax.grid(True, which='major', linestyle='-.', axis='y' )
ax.set_xticklabels(paises['country'],rotation=90)
ax.figure.set_figheight(6)
ax.figure.set_figwidth(15)
```
